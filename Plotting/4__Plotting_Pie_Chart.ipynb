```python
# import required libraries
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# data from Ohio state, leading causes of death on the state from year 2012
# cause of death
cause = 'Chronic Diseases','Unintentional','Alzheimers','Infuenza and Pneumonia','Stepsis','Others'
# percentile
percentile = [62,5,4,2,1,26]
```

```python
# set the pie chart plot properties
# set figure size
plt.figure(figsize = (10,10))
# explode the largest pie in the dataset
explode_1 = (0.05,0,0,0,0,0)
# set pie properties
plt.pie(percentile,labels=cause,explode=explode_1,autopct ='% 1.1f %%',shadow = True, startangle = 90)
# set axis equal to draw pie as circle
plt.axis('equal')
# set title of the pie chart
plt.title('Ohio State - 2012 : Leading Causes of Death')
# show the plot
plt.show()
```
