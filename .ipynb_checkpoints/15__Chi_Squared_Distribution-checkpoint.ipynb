# CHI-SQUARED-DISTRIBUTION
Its is a continuous probability distribution.
Mainly used for hypothesis testing, Chi squre goodness field, or Chi squre test of independence.
It is usually defined by Degree of independence.

```python
import numpy as np
data1 = np.random.chisquare(df = 1, size = 1000)
data2 = np.random.chisquare(df = 2, size = 1000)
data3 = np.random.chisquare(df = 3, size = 1000)
```

```python
print(data1[:10])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.distplot(data1, hist = False, label = 'dof 1')
sns.distplot(data2, hist = False, label = 'dof 2')
sns.distplot(data3, hist = False, label = 'dof 3')
plt.legend()
```

```python
from scipy.stats import chi2_contingency
```

```python
data = [[10,20,30],[6,9,17]]  # null hypothesis is that there is no relation between 2 variables that we are testing for
                              # alternative hypothesis is that is a correlation between 2 variables that we are testing for
stat,p_value,dof,chi_array = chi2_contingency(data)
```

```python
p_value # so the output of p_ value is greater tha 0.5 than the null hypothesis is rejected and alternative hypothesis is accepted
```

```python
data = [[10,20,30],[9,1,8]]  # null hypothesis is that there is no relation between 2 variables that we are testing for
                              # alternative hypothesis is that is a correlation between 2 variables that we are testing for
stat,p_value,dof,chi_array = chi2_contingency(data)
```

```python
p_value # so the output of p_ value is greater tha 0.5 than the alternative hypothesis is rejected and null hypothesis is accepted
```
