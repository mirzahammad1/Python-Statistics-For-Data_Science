```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import f
x = np.linspace(0,4.5,1000)
data_1 = f(1,1,0)
data_2 = f(5,8,0)
data_3 = f(4,4,0)
data_4 = f(100,200,0)
plt.figure(figsize = (12,6) , dpi = 150)
plt.plot(x ,data_1.pdf(x), label = '1,1')
plt.plot(x ,data_2.pdf(x), label = '5,8')
plt.plot(x ,data_3.pdf(x), label = '4,4')
plt.plot(x ,data_4.pdf(x), label = '100,200')
plt.legend()
```

```python
Null Hypothesis & Variances are equal
Alternative hypothesis are not equal 
```

```python
import random
random.seed(20)
```

```python
x = np.array([random.gauss(100,15) for x in range(20)])
y = np.array([random.gauss(100,15) for x in range(20)])
```

```python
f_test_stat = np.var(x,ddof=1)/np.var(y,ddof=1)  # ddof is one to get the sample variance instead of population variance
```

```python
dfn = x.size - 1
dfd = y.size - 1
```

```python
p_value = 1 - f.cdf(f_test_stat,dfn,dfd)
p_value       # the output will be less than 0.05 which means our null hypothesis is true
              # which mean there is a relation between these variables OR the variance are equal
```

```python
x = np.array([random.gauss(100,15) for x in range(20)])
y = np.array([random.gauss(200,15) for x in range(20)])
f_test_stat = np.var(x,ddof=1)/np.var(y,ddof=1)
dfn = x.size - 1
dfd = y.size - 1
p_value = 1 - f.cdf(f_test_stat,dfn,dfd)
p_value       # the output will be greater than 0.05 which means our alternative hypothesis is true
              # which mean the variance are not equal
```
