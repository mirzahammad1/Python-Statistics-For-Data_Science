# Problem Statement
suppose you are going for a long drive. The rate of occurrences of good restaurants in range of 10 miles is 2. In other words, the mean number of occurrences of restaurants in range of 10 miles is 2. What is the probability that 0,1,2,3,4 or 5 reastaurants will show up in the next 10 miles. 

```python
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python
# Random variable representing number of resturant
# Mean number of occurrences of resturants in 10 miles is 2
X = [0,1,2,3,4,5]
lmbda = 2
```

```python
# Probability values
poisson_pd = poisson.pmf(X,lmbda)
```

```python
# plot the probability distribution

fig, ax = plt.subplots (1, 1, figsize = (9, 7))
ax.plot(X, poisson_pd, 'bo', ms = 8, label = 'poisson pmf')
plt.ylabel("Probability", fontsize = "18")
plt.xlabel("X = No. of Resturants", fontsize = "18")
plt.title("Poisson Distribution -- No. od Resturants vs Probability", fontsize = "18")
ax.vlines(X, 0, poisson_pd, colors = 'b', lw = 5 , alpha = 0.5)
```
